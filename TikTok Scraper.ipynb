```python
from ensembledata.api import EDClient
```

```python
API_TOKEN = "VNxUBwtsw7Zad9rw"
client = EDClient(API_TOKEN)
```

```python
username = "itsgoneviral"
result = client.tiktok.user_posts_from_username(username=username, depth=1)
```

```python
import pandas as pd

df_raw = pd.DataFrame(result.data)
df_raw.head()
```

```python
df_expanded = pd.json_normalize(df_raw['data'])
df_expanded.head()
```

```python
df_clean = df_expanded[[
    'desc',
    'statistics.play_count',
    'statistics.digg_count',
    'statistics.comment_count',
    'statistics.share_count',
    'statistics.collect_count',
    'video.duration',
    'create_time'
]]
```

```python
df_clean.rename(columns={
    'desc': 'description',
    'statistics.play_count': 'views',
    'statistics.digg_count': 'likes',
    'statistics.comment_count': 'comments',
    'statistics.share_count': 'shares',
    'statistics.collect_count': 'favorites',
    'video.duration': 'duration_secs',
    'create_time': 'timestamp'
}, inplace=True)
```

```python
# Convert timestamp to datetime
df_clean['timestamp'] = pd.to_datetime(df_clean['timestamp'], unit='s')
```

```python
df_clean.head(10)
```

```python
df_clean.to_csv("IGVTiktok_data.csv", index=False)
```
